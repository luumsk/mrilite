# Demonstration for the use of `mrilite`

```python
from mrilite import (
    load_nifti,
    zscore,
    show_slice,
    show_slice_with_mask,
    show_overlay,
    show_modalities
)
```

```python
# Example data is from the Individual Brain Charting dataset (OpenNeuro ds000244, CC0 license), patient 1
img_path = "img/T1W.nii.gz"
seg_path = "img/consensus_gt.nii.gz"

img = load_nifti(img_path)
seg = load_nifti(seg_path)

# show_slice_with_overlay(img, seg, axis=2, index=80, alpha=0.4)
```

```python
img.describe()
```

```python
fig, axes = show_slice_with_mask(img, seg, axis=1, index=100)
```

```python
fig, ax = show_overlay(img, seg, axis=1, index=100)
```

```python
volumnes = {
    "T1W": load_nifti("img/T1W.nii.gz"),
    "T2W": load_nifti("img/T2W.nii.gz"),
    "FLAIR": load_nifti("img/FLAIR.nii.gz"),
}
show_modalities(volumnes, axis=1, index=100)
```
